# Approximating Optimized Model Coefficients

```python
import numpy as np
import pandas as pd
import math
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import pylab as pl
from IPython.html.widgets import *
```

```python
PendData = np.loadtxt('data/pendData.csv', skiprows=1, delimiter=',')
PendFrame = pd.DataFrame(PendData,columns=['time(sec)','angular position(theta)'])
```

```python
def sine_wave(time,amp,freq,phi,vertShift,damp):
    return ((math.exp(-1*damp*time))*(amp*math.sin(2*math.pi*freq*time-phi)+vertShift))
```

Where: <br>
   t = Time, in seconds<br>
   A = Peak Amplitude<br>
   f = Frequency, in Hz<br>
   &Phi; = Phase Shift<br>
   B = Vertical shift

```python
def sine_values(amp,freq,phi,vertShift,damp):
    PendFrame['SineValues']= PendFrame.apply(lambda row: sine_wave(row['time(sec)'],amp,freq,phi,vertShift,damp),axis=1)
```

```python
def sine_values2(amp,freq,phi,vertShift,damp):
    sineValues = []
    for time in PendFrame['time(sec)']:
        sineValues.append(sine_wave(time,amp,freq,phi,vertShift,damp))
    ts = pd.Series(sineValues, index = PendFrame['time(sec)'])
    return ts
```

```python
def model(amp,freq,phi,vertShift,damp):
    
    ts =sine_values2(amp,freq,phi,vertShift,damp)

    plt.figure(figsize=(18,6))

    #TSPlot = PendFrame.plot(x='time(sec)',y='angular position(theta)',ylim=(-22,20),linewidth=.7, c='red')
    TSPlot = pendTS.plot(ylim=(-22,22),linewidth=.7, c='red')
    ts.plot(linewidth=.7, c='blue')
    
    #PendFrame.plot(x='time(sec)',y='SineValues',ylim=(-22,22),linewidth=.7, c='blue')
    #plt.axis([0.0,36.0,-22.0,22.0])
    TSPlot.set_axis_bgcolor('white')
    TSPlot.set_ylabel('pendulum angular position (degrees)', fontsize=16)
    TSPlot.set_xlabel('time (seconds)',fontsize=16)
    TSPlot.set_title('Approximating Model Coefficients',fontsize=20)
    TSPlot.legend(['Fit Data','Model','Observed Data Points (time,theta)'],1)
    plt.tick_params(axis='both', labelsize=15)
    
    SSE = np.sum((PendFrame['angular position(theta)']-PendFrame['SineValues'])**2)
    print("SSE:",SSE)
```

```python
interact(model,amp=(15.0,25.0,.001),freq=(1.0,2.0,.001),phi=(-1.0,1,.001),vertShift=(-2.0,2.0,.001),damp=(.001,.1,.001));
```

```python
def normal_2d(n, mux, muy, sigmax, sigmay, corr):
    mean = [mux, muy]
    cov = [[sigmax**2, corr*sigmax*sigmay],[corr*sigmax*sigmay,sigmay**2]]
    d = np.random.multivariate_normal(mean, cov, n)
    return d[:,0], d[:,1]
```

```python
def plot_normal_2d(n, mux, muy, sigmax, sigmay, corr):
    x, y = normal_2d(n, mux, muy, sigmax, sigmay, corr)
    plt.scatter(x, y, s=100, alpha=0.6)
    plt.axis([-10.0,10.0,-10.0,10.0])
```

```python
interact(plot_normal_2d, n=(10,10000,10), mux=(-5.0,5.0,0.1), muy=(-5.0,5.0,0.1),
         sigmax=(0.01,5.0,0.00001), sigmay=(0.01,5.0,0.00001), corr=(-0.99,0.99,0.000001));
```
