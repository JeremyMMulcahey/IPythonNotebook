### 9.1.1 Converting x and y values to angular position (&Theta;)

```python
#read the csv file.
PendData = np.loadtxt('data/pend_data.csv', skiprows=1, delimiter=',')

#convert the NumPy structure into a DataFrame.
PendFrame = pd.DataFrame(PendData,columns=['time(sec)','x(pixels)',
                                           'y(pixels)'])
```

Using the provided dataset and center of the circle, we can use the following formula to convert the x and y values to the angular position of the pendulum at time t (theta).

<img src="images/angular.PNG", height="50%", width="50%">

<br>

The new theta column will enable us to analyze a time series in the following format:<br>
<img src="images/timeSeries.PNG", height="20%", width="20%">

The following code declares a function to calculate theta, then applies the function to the existing x and y columns of the DataFrame to create a new column named theta.

```python
#function takes y and x, then returns the solution to the above equation.
def calc_theta(y,x):
    return ((180/math.pi)*math.atan2((y-1152.57606607263),
                                     (x-394.773399557239)))+90

#the left side creates a new column in the DataFrame named "theta".
#the right side uses information from the existing columns and 
#the calc_theta function above to populate the "theta" column.
PendFrame['theta']= PendFrame.apply(lambda row: 
                                    calc_theta(row['y(pixels)'],
                                               row['x(pixels)']), axis=1)
PendFrame.head()
```

In the code box above, populating the theta column is done row by row by using the <b>.apply()</b> function and the <b>lambda row:</b> argument, which specifies to use the <b>calc_theta</b> function on the x and y values on each row.

### 9.1.4 Modeling a Damped Sine Wave

In Section 9.1.2, the time series plot showed a sine wave that decreased in amplitude as time progressed. To model this, use the formula for a damped sine wave provided below:<br>

<img src="images/dampedSin.PNG", height = 40%, width = 40%>

Where: <br>
   t = Time, in seconds<br>
   A = Peak Amplitude<br>
   f = Frequency, in Hz<br>
   &Phi; = Phase Shift<br>
   B = Vertical shift

As demonstrated in Section 9.1.1, a column will be added to the DataFrame using the apply function and a function to calculate the formula the above.

```python
#function for the sine wave of interest.
def sine_wave(time,amp,freq,phi,vertShift,damp):
    return ((math.exp(-1*damp*time))*
            (amp*math.sin(2*math.pi*freq*time-phi)
             +vertShift))
```

The calculation for SSE has been added to the code below. The individual variable values can be changed, and the code re-executed, to minimize the SSE by trial and error. The lowest SSE value will determine the best estimate for the model.

```python
#changing these values and executing this code box after each change 
#will alter the SSE value. 
amp = 19.649
freq = 1.26082
phi = .18205
vertShift = -.4705
damp = .02695

#creates a new column in the DataFrame and populates it
#with sine wave values.
PendFrame['SineValues']= PendFrame.apply(lambda row: 
                                         sine_wave(row['time(sec)'],
                                                   amp,freq,phi,
                                                   vertShift,damp),
                                         axis=1)

#print sum of squared errors to see if error is decreasing.
SSE = np.sum((PendFrame['theta']-PendFrame['SineValues'])**2)
print("SSE:",SSE)
```

The values provided in the code above are already precise approximations of the parameters. Changing them might yield a slightly lower SSE, but it should not be drastically different than the SSE printed above.

Overlay the model on the time series plot from Section 9.1.2.

```python
#open a figure window of size 18x6.
plt.figure(figsize=(18,6))

#plot observed points.
plt.scatter(PendFrame['time(sec)'],PendFrame['theta'], s=8, c='black')

#plot observed line.
TSPlot = PendFrame.plot(x='time(sec)',y='theta',ylim=(-22,20),
                        linewidth=.7, c='red')

#overlay model line.
PendFrame.plot(x='time(sec)',y='SineValues',ylim=(-22,22),
               linewidth=.7, c='blue')

#change background to white.
TSPlot.set_axis_bgcolor('white')

#set x and y labels, title, and adjust font sizes accordingly.
TSPlot.set_ylabel('pendulum angular position (degrees)', fontsize=16)
TSPlot.set_xlabel('time (seconds)',fontsize=16)
TSPlot.set_title('Pendulum Position Time Series: Model and Observed Fit',
                 fontsize=20)
TSPlot.legend(['Fit Data','Model','Observed Data Points (time,theta)'],1)
plt.tick_params(axis='both', labelsize=15)
```

Plot the resdiuals to see if a pattern remains.

```python
#residuals are (observed values at ti) - (values from model at ti).
def calc_resid(obs,pred):
    return obs-pred

#creates a new column in the DataFrame and populates it with residuals.
PendFrame['Residuals']= PendFrame.apply(lambda row:
                                        calc_resid(row['theta'],
                                                   row['SineValues']),
                                        axis=1)
```

```python
#open a figure window.
plt.figure(figsize=(18,6))

#plot the scatterplot first to keep the markers in the foreground.
plt.scatter(PendFrame['time(sec)'],PendFrame['Residuals'], s=8, 
            c='black')

#built-in DataFrame function to plot time as x and theta as y,
#with custom y limits, line width, and color.
TSPlot = PendFrame.plot(x='time(sec)',y='Residuals',linewidth=.7, 
                        c='green')

#set x and y labels, title, and adjust font sizes accordingly.
TSPlot.set_ylabel('Residuals', fontsize=16)
TSPlot.set_xlabel('Time (seconds)',fontsize=16)
TSPlot.set_title('Residuals as time progresses',fontsize=20)
plt.tick_params(axis='both', labelsize=15)
```

From the residual plot, it appears the model can be improved to account for the left over stochastic component in the residuals.

With a dataset that is almost pure noise, the simple moving average does not produce an exploitable pattern in the Geiger counter data.

The code below contains a demonstration of how to write a centered moving average (CMA).

```python
#function for CMA.
def CMA(vec, n):
    c = n/2
    cma = np.zeros([16384,1])  #creates an empty array to store values.
    for i in range(c, len(vec)-c):    #values from n/2 to 16384-n/2.
        window = vec[i-c:i+c+1]       #taking average over these values.
        cma[i] = (np.mean(window))    #store average in array.
    cma[:c] = None                    #fill first n/2 values with None.
    cma[(len(vec)-c):] = None         #fill last n/2 values with None.
    return cma
```

```python
#add column to the DataFrame.
gcFrame['CMA'] = CMA(gcFrame.Detections,24)
```

### 9.3.3 Custom time series slices and weighted intervals

The next step to create a single time series with constant changes in time and price weighted by quantity is to manually generate the 5 minute time intervals. The 5 minute time intervals will be used to slice the original data by time indices. 

```python
#generate time values every 5 minutes on October 13th, 2014.
intervals = pd.date_range('10/13/2014', periods=270, freq='5min')

#create a time series of the time values.
intervals2 = pd.Series(intervals,index =intervals)

#slice the generated times to match the time window provided by NetFonds.
indexForWeights = intervals2['2014-10-13 15:30:00':'2014-10-13 22:00:00']

#create an index for the 79 time values.
index = np.arange(79)

#create a time series with integer indexing and times are the
#indexed values.
reindex = indexForWeights.reindex(index)

#manually add the times as the indexed values.
reindex[0:] = indexForWeights
reindex.head()
```

The function below will utilize the price and quantity time series objects and the time series object containing the desired time intervals to create a single time series that incorporates all the data.

```python
#function to create five minute weighted windows.
def weighted_fiveMinInt(quantityTS,priceTS,tsOfIntervals):
    
    #the weight of the interval is the number of stocks bought
    #at the respective selling price.
    weights = quantityTS*priceTS
    fiveMinIntervals = np.zeros(len(tsOfIntervals)-1)
    
    for i in range(len(tsOfIntervals)-1):
        #using time series slicing to sum all values over the 
        #time interval.
        weightSlice = sum(weights[tsOfIntervals[i]:tsOfIntervals[i+1]])
        quantSlice = sum(quantityTS[tsOfIntervals[i]:tsOfIntervals[i+1]])
        
        #total weighted price of the interval.
        newPrice = weightSlice/quantSlice
        fiveMinIntervals[i] = newPrice
    return fiveMinIntervals
```

<br>

Use the function to combine the data.

```python
weightedWindows = weighted_fiveMinInt(NFLXTSquant,NFLXTSprice,reindex)
```

Create a single time series for the data.

```python
weightedWindowsTS = pd.Series(weightedWindows, 
                              index=indexForWeights[1:])
```

Plot the smoothed/weighted time series against the original stock ticker data.
