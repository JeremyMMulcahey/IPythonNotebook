## 8.2 Streaming Data from USA.gov

USA.gov describes the data as, "We provide a raw pub/sub feed of data created any time anyone clicks on a 1.USA.gov URL. The pub/sub endpoint responds to http requests for any 1.USA.gov URL and returns a stream of JSON entries, one per line, that represent real-time clicks."

```python
#create a variable for the datastream url.
url = "http://developer.usa.gov/1usagov"
```

```python
#this function gets the data from the provided url.
r = requests.get(url, stream=True)

#n specifies the number of clicks we want to collect data on.
n = 500
data = []

#looks at each line of the request individually and adds the information
#to the list above ("data").
for i, line in enumerate(r.iter_lines()):
    data.append(line)
    
    #under normal circumstances using break to stop a loop is not
    #recommended. however, it is useful for ending the live 
    #datastream.
    if i > n:
        break
```

```python
#load the json lines from the list named data.
jdata = [json.loads(item) for item in data[1:]]
```

```python
#create a DataFrame of the click data.
USAGovFrame = pd.DataFrame(jdata)
```

Now that the data is in a DataFrame, we can use techniques from the previous chapter to view and analyze the data.

```python
USAGovFrame.head()
```

After inspecting the DataFrame, several of the variables provided by the datastream contain mostly NaN values. To simplify the DataFrame, drop the columns that are mostly NaN and/or don't apply to any potential analysis we might perform.

```python
#ckw and dp are not consistently collected, especially with 
#smaller sample sizes this will check for them and drop them 
#if they are present.
if 'ckw' and 'dp' in USAGovFrame.columns:
    USAGovFrame.drop(['_heartbeat_','_id','al','ckw','nk',
                      'g','h','kw','hc','dp'],inplace=True,
                     axis=1)
else:
    USAGovFrame.drop(['_heartbeat_','_id','al','nk','g',
                      'h','kw','hc'],inplace=True, axis=1)
```

As stated in the comment above, some variables are not always present. If you wish to use this exact code to trim your DataFrame, then make sure n is at least 100. If an error is encountered, check the variables present in your DataFrame and adjust the drop function accordingly.

```python
#add user friendly names for the remaining variables.
USAGovFrame.columns=['User_Agent','Country_Code',
                     'Geo_city_name','Geo_Region',
                     'Short_url_Cname','Encoding_user_login',
                     '[Latitude,Longitude]','Referring_URL',
                     'Timestamp','Timezone','Long_URL']
```

## 8.3 Plotting on a world map

The coordinates need to be examined, and possibly manipulated, before being able to plot them on a map.

The type function can be used to determine the type/mode of an object.

```python
#check the type of data at the first index.
type(USAGovFrame['[Latitude,Longitude]'][0])
```

On your own, read about unzipping. Under normal circumstances, separating and storing information from a list (unzipping) into individual lists is relatively trivial. In this instance, I encountered an error that I chose to address with a function.

```python
#function for unpacking a list in each row of a column into 
#two separate lists.
def lat_lon(column):
    
    #create two empty lists.
    lat=[]
    lon=[]
    
    #at each row in the column, add first index to lat, 
    #second index to lon.
    for i in column:
        lat.append(i[0])
        lon.append(i[1])
        
    #return the new lists.
    return lat,lon
```

The function in the code box above will be called in the code box for plotting the map, below this cell.

```python
#open a figure of size 20x10.
plt.figure(figsize=(20,10))

# lon_0 is central longitude of projection.
# resolution = 'c' means use crude resolution coastlines.
map = Basemap(projection='kav7',lon_0=0,resolution='c')

#bluemarble is a built in function to basemap. 
map.bluemarble()

#drop the NaN values from the data frame.
coords = USAGovFrame['[Latitude,Longitude]'].dropna()

#run the function from the previous code box.
lat, lon = lat_lon(coords)

#plot the points on the bluemarble basemap.
# '.' is the marker type, c is the color.
x,y = map(lon, lat)
map.plot(x,y,'.', c='red',markersize=12)
plt.title("USA.Gov Site Traffic", fontsize = 20);
```
