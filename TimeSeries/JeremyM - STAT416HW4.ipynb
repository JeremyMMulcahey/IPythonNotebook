Using the variances of the models:

```python
def AIC(k,var):
    return 2*k + math.log(var)

def AICc(k,n,AIC):
    return AIC + (2*k*(k+1))/(n-k-1)

def BIC(k,n,var):
    return k*math.log(n)-2*math.log(var)
```

```python
arima212AIC = AIC(4,24.464)
arima212AICc = AICc(4,411,arima212AIC)
arima112AIC = AIC(3,25.101)
arima112AICc = AICc(3,411,arima112AIC)
arima212BIC = BIC(4,411,24.464)
arima112BIC = BIC(3,411,25.101) 
print("ARIMA(2,1,2)x(0,1,0)100 AICc:",arima212AICc)
print("ARIMA(1,1,2)x(0,1,0)100 AICc:",arima112AICc)
print("ARIMA(2,1,2)x(0,1,0)100 BIC:",arima212BIC)
print("ARIMA(1,1,2)x(0,1,0)100 BIC:",arima112BIC)
```
