```python
import numpy as np
import pandas as pd
import math
import matplotlib
import matplotlib.pyplot as plt
import statsmodels.tsa as tsa
from datetime import datetime
import time
import matplotlib.dates as mdates
from statsmodels.formula.api import ols
import cmath
%matplotlib inline
```

```python
sunspot = pd.read_csv('data/sunspot.csv')
co2 = pd.read_csv('data/weeklyCO2.csv')
```

```python
#convert the dates and times to something that can be graphed as a scatter plot
sunspot['date'] = [datetime.strptime(t, "%m/%d/%Y").date() for t in sunspot['date']]
sunspot['scatter'] = mdates.date2num(sunspot['date'])
```

```python
#convert the dates and times to something that can be graphed as a scatter plot
co2['date'] = [datetime.strptime(t, "%m/%d/%Y").date() for t in co2['date']]
co2['scatter'] = mdates.date2num(co2['date'])
```

```python
ssTS = pd.Series(sunspot['sunspot'], index = sunspot['scatter'])
co2TS = pd.Series(co2['average'], index = sunspot['scatter'])
```

```python
ssTS[0:] = sunspot['sunspot']
co2TS[0:] = co2['average']
```

```python
sslm = ols('sunspot ~ scatter',sunspot).fit()
sslmTS = pd.Series(sslm.fittedvalues, index = sunspot['scatter'])
```

```python
colm = ols('average ~ scatter',co2).fit()
colmTS = pd.Series(colm.fittedvalues, index = co2['scatter'])
```

```python
sslmTS[0:] = sslm.fittedvalues
colmTS[0:] = colm.fittedvalues
```

```python
fig = plt.figure(figsize=(20,8))
plt.scatter(sunspot['scatter'],sunspot['sunspot'], s=5,label='Observed Data Points', c='black')
coTS2 = co2TS.plot(secondary_y = True, linewidth = 2, c='blue', label = 'C02',legend=True)
TSPlot = ssTS.plot(linewidth=1, c='green', label='Sunspot Number',ylim=(0,300),legend=True)
sslmTS.plot(linewidth=2, c='orange', label='Sunspot Fit', legend=True)
coTS2.plot(sunspot['scatter'],colm.fittedvalues,linewidth=2, c='purple', label = 'CO2 Fit')

xlabels = []
xticks = []
for i in range(51):
    xlabels.append(sunspot['date'][i*40])   
    xticks.append(sunspot['date'][i*40])

TSPlot.set_xticklabels(xlabels, rotation=90, fontsize=15) 
TSPlot.set_xticks(xticks)
TSPlot.set_ylabel('Sunspot Number', fontsize=15)
TSPlot.set_xlabel('Date (y/m/d)', fontsize=15)
coTS2.set_ylabel('Atmospheric CO2 concentration', fontsize=15)

#set x and y labels, title, and adjust thier sizes according
TSPlot.set_title('Weekly average atmospheric CO2 and sunspot numbers from 1974-2012',fontsize=20)
plt.tick_params(axis='both', labelsize=15)
TSPlot.tick_params(axis='y', labelsize=15)
```

```python
def hanning(col):
    hanning=[]
    for j in np.arange(len(col)):
        calc = .5*(1-(math.cos((2*math.pi*(j+1))/float(2048))))
        hanning.append(calc)
    return hanning

padFrame['sshanning']= hanning(padFrame['sunspot'])*padFrame['sunspot']
padFrame['co2hanning']= hanning(padFrame['average'])*padFrame['average']
```

```python
padFrame['scatter'] = mdates.date2num(padFrame['date'])
```

```python
fig = plt.figure(figsize=(20,8))
plt.scatter(padFrame['scatter'],padFrame['sshanning'], s=5,label='Observed Data Points', c='red')
coTS2 = padFrame.plot(secondary_y = True, y='co2hanning', x='scatter', linewidth = 1, c='black', label = 'C02',legend=True, ylim=(-6,12))
plt.scatter(padFrame['scatter'],padFrame['co2hanning'], s=5,label='Observed Data Points', c='lightblue')
TSPlot = padFrame.plot(y='sshanning',x='scatter',linewidth=1.5, c='lightblue', label='Sunspot Number',ylim=(-100,200),legend=True)


xlabels = []
xticks = []
for i in range(52):
    xlabels.append(padFrame['date'][i*40])   
    xticks.append(padFrame['date'][i*40])

TSPlot.set_xticklabels(xlabels, rotation=90, fontsize=15) 
TSPlot.set_xticks(xticks)
TSPlot.set_ylabel('Sunspot Number', fontsize=15)
TSPlot.set_xlabel('Date (y/m/d)', fontsize=15)
coTS2.set_ylabel('Atmospheric CO2 concentration', fontsize=15)

#set x and y labels, title, and adjust thier sizes according
TSPlot.set_title('De-Trended, padded, and Hanning Windowed Weekly avg. atmos. CO2 and sunspot numbers from 1974-2013',fontsize=20)
plt.tick_params(axis='both', labelsize=15)
TSPlot.tick_params(axis='y', labelsize=15)
```

```python
ssfft = np.fft.fft(padFrame['sunspot'])
co2fft = np.fft.fft(padFrame['average'])
```

```python
dtFFTFrame = pd.DataFrame(ssfft)
dtFFTFrame['co2fft'] = co2fft
```

```python
dtFFTFrame.columns=['ssfft','co2fft']
```

```python
def imabs(complexNum):
    return 2*abs(complexNum)/float(2048)
```

```python
dtFFTFrame['peakAmpss']= dtFFTFrame.apply(lambda row: imabs(row['ssfft']), axis=1)
dtFFTFrame['peakAmpco2']= dtFFTFrame.apply(lambda row: imabs(row['co2fft']), axis=1)
```

```python
dtFFTFrame['peakAmpss'][0] = 0
dtFFTFrame['peakAmpss'][1023] = abs(ssfft[1023])
```

```python
dtFFTFrame['peakAmpco2'][0] = 0
dtFFTFrame['peakAmpco2'][1023] = abs(co2fft[1023])
```

```python
dtFFTFrame['k'] = np.arange(len(ssfft))
```

```python
dtFFTFrame['freq'] = dtFFTFrame['k']/(2048*7)
```

```python
fig = plt.figure(figsize=(20,8))
plt.scatter(dtFFTFrame['freq'],dtFFTFrame['peakAmpss'], s=15,label='Observed Data Points', c='red')
coTS2 = dtFFTFrame.plot(secondary_y = True, y='peakAmpco2', x='freq', linewidth = 2, c='black', label = 'C02',legend=True, ylim=(0,2),xlim=(0,.01))
plt.scatter(dtFFTFrame['freq'],dtFFTFrame['peakAmpco2'], s=25,label='Observed Data Points', c='blue')
TSPlot = dtFFTFrame.plot(y='peakAmpss',x='freq',linewidth=2, c='blue', label='Sunspot Number',ylim=(0,55),legend=True,xlim=(0,.01))

TSPlot.set_ylabel('Spectral Density Peak Amp SSN', fontsize=15)
TSPlot.set_xlabel('Frequency (cycles per day)', fontsize=15)
coTS2.set_ylabel('Spectral Density Atmospheric CO2 concentration', fontsize=15)

TSPlot.set_title('De-trended Periodogram',fontsize=20)
plt.tick_params(axis='both', labelsize=15)
TSPlot.tick_params(axis='y', labelsize=15)
TSPlot.legend(['PSD Detreneded SSN fit','PSD Detreneded SSN data'],loc=9)
coTS2.legend(['PSD Detreneded CO2 fit','PSD Detreneded C02 data'],loc=10);
```

```python
def low_pass_filter(freq,fft):
    if freq < .001:
        return fft
    else:
        return 0
    
dtFFTFrame['lowpassFilterss'] = dtFFTFrame.apply(lambda row: low_pass_filter(row['freq'],row['ssfft']), axis=1)
dtFFTFrame['lowpassFilterco2'] = dtFFTFrame.apply(lambda row: low_pass_filter(row['freq'],row['co2fft']), axis=1) 
```

```python
dtFFTFrame['lowpassIFFTss'] = np.fft.ifft(dtFFTFrame['lowpassFilterss'])/15
dtFFTFrame['lowpassIFFTco2'] = np.fft.ifft(dtFFTFrame['lowpassFilterco2'])/.015
```

```python
dtFFTFrame['scatter']=padFrame['scatter']
```

```python
fig = plt.figure(figsize=(20,8))
coTS2 = dtFFTFrame.plot(secondary_y = True, y='lowpassIFFTss', x='scatter', linewidth = 4, c='blue', label = 'Low Pass SSN',legend=True, ylim=(-4,4))
TSPlot = dtFFTFrame.plot(y='lowpassIFFTco2',x='scatter',linewidth=4, c='red', label='Low Pass CO2',ylim=(-100,150),legend=True)

xlabels = []
xticks = []
for i in range(52):
    xlabels.append(padFrame['date'][i*40])   
    xticks.append(padFrame['date'][i*40])

TSPlot.set_xticklabels(xlabels, rotation=90, fontsize=15) 
TSPlot.set_xticks(xticks)
TSPlot.set_ylabel('Sunspot Number', fontsize=15)
TSPlot.set_xlabel('Date (y/m/d)', fontsize=15)
coTS2.set_ylabel('Atmospheric CO2 concentration', fontsize=15)

#set x and y labels, title, and adjust thier sizes according
plt.title('Low-pass filter on padded and detreneded data series', fontsize=20)
plt.tick_params(axis='both', labelsize=15)
TSPlot.tick_params(axis='y', labelsize=15)
coTS2.legend(loc=2)
TSPlot.legend();
```

```python
def band_pass_filter(freq,fft,k):
    if 4<= k <= 6:
        return fft
    else:
        return 0
    
dtFFTFrame['bandpassFilterss'] = dtFFTFrame.apply(lambda row: band_pass_filter(row['freq'],row['ssfft'],row['k']), axis=1)
dtFFTFrame['bandpassFilterco2'] = dtFFTFrame.apply(lambda row: band_pass_filter(row['freq'],row['co2fft'],row['k']), axis=1) 
```

```python
dtFFTFrame['bandpassIFFTss'] = np.fft.ifft(dtFFTFrame['bandpassFilterss'])/75
dtFFTFrame['bandpassIFFTco2'] = np.fft.ifft(dtFFTFrame['bandpassFilterco2'])/.005
```

```python
fig = plt.figure(figsize=(20,8))
coTS2 = dtFFTFrame.plot(secondary_y = True, y='bandpassIFFTss', x='scatter', linewidth = 4, c='blue', label = 'Band Pass SSN',legend=True, ylim=(-.7,.8))
TSPlot = dtFFTFrame.plot(y='bandpassIFFTco2',x='scatter',linewidth=4, c='red', label='Band Pass CO2',ylim=(-100,100),legend=True)

xlabels = []
xticks = []
for i in range(52):
    xlabels.append(padFrame['date'][i*40])   
    xticks.append(padFrame['date'][i*40])

TSPlot.set_xticklabels(xlabels, rotation=90, fontsize=15) 
TSPlot.set_xticks(xticks)
TSPlot.set_ylabel('Sunspot Number', fontsize=15)
TSPlot.set_xlabel('Date (y/m/d)', fontsize=15)
coTS2.set_ylabel('Atmospheric CO2 concentration', fontsize=15)

#set x and y labels, title, and adjust thier sizes according
plt.title('Low-pass filter on padded and detreneded data series', fontsize=20)
plt.tick_params(axis='both', labelsize=15)
TSPlot.tick_params(axis='y', labelsize=15)
coTS2.legend(loc=2)
TSPlot.legend();
```
