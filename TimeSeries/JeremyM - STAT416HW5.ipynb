```python
import numpy as np
import pandas as pd
import math
import matplotlib
import matplotlib.pyplot as plt
import statsmodels.tsa as tsa
from datetime import datetime
import time
import matplotlib.dates as mdates
from statsmodels.formula.api import ols
%matplotlib inline
```

```python
sunspot = pd.read_csv('data/sunspot.csv')
co2 = pd.read_csv('data/weeklyCO2.csv')
```

```python
#convert the dates and times to something that can be graphed as a scatter plot
sunspot['date'] = [datetime.strptime(t, "%m/%d/%Y").date() for t in sunspot['date']]
sunspot['scatter'] = mdates.date2num(sunspot['date'])
```

```python
#convert the dates and times to something that can be graphed as a scatter plot
co2['date'] = [datetime.strptime(t, "%m/%d/%Y").date() for t in co2['date']]
co2['scatter'] = mdates.date2num(co2['date'])
```

```python
ssTS = pd.Series(sunspot['sunspot'], index = sunspot['scatter'])
co2TS = pd.Series(co2['average'], index = sunspot['scatter'])
```

```python
ssTS[0:] = sunspot['sunspot']
co2TS[0:] = co2['average']
```

```python
sslm = ols('sunspot ~ scatter',sunspot).fit()
sslmTS = pd.Series(sslm.fittedvalues, index = sunspot['scatter'])
```

```python
colm = ols('average ~ scatter',co2).fit()
colmTS = pd.Series(colm.fittedvalues, index = co2['scatter'])
```

```python
sslmTS[0:] = sslm.fittedvalues
colmTS[0:] = colm.fittedvalues
```

```python
fig = plt.figure(figsize=(20,8))
plt.scatter(sunspot['scatter'],sunspot['sunspot'], s=5,label='Observed Data Points', c='black')
coTS2 = co2TS.plot(secondary_y = True, linewidth = 2, c='blue', label = 'C02',legend=True)
TSPlot = ssTS.plot(linewidth=1, c='green', label='Sunspot Number',ylim=(0,300),legend=True)
sslmTS.plot(linewidth=2, c='orange', label='Sunspot Fit', legend=True)
coTS2.plot(sunspot['scatter'],colm.fittedvalues,linewidth=2, c='purple', label = 'CO2 Fit')

xlabels = []
xticks = []
for i in range(51):
    xlabels.append(sunspot['date'][i*40])   
    xticks.append(sunspot['date'][i*40])

TSPlot.set_xticklabels(xlabels, rotation=90, fontsize=15) 
TSPlot.set_xticks(xticks)
TSPlot.set_ylabel('Sunspot Number', fontsize=15)
TSPlot.set_xlabel('Date (y/m/d)', fontsize=15)
coTS2.set_ylabel('Atmospheric CO2 concentration', fontsize=15)

#set x and y labels, title, and adjust thier sizes according
TSPlot.set_title('Weekly average atmospheric CO2 and sunspot numbers from 1974-2012',fontsize=20)
plt.tick_params(axis='both', labelsize=15)
TSPlot.tick_params(axis='y', labelsize=15)
```

```python
ssDT = ssTS - sslmTS
co2DT = co2TS - colmTS
```
