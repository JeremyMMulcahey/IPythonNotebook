## Introduction

This chapter focuses on exponentially smoothing methods to make forecast predictions for stocks, assessing autocorrelation, and creating an MA model for the previously introduced geiger counter data.

## Datasets

Stock ticker data for Netflix, from NetFonds, was obtained for October 13, 2014. The data is converted to 5 minute intervals and weighted by quantity of stocks sold in the 5 minute period before employing a single and double exponential smoother to predict the stock price.

```python
from IPython.display import Image
import numpy as np
import pandas as pd
import math
import matplotlib
import matplotlib.pyplot as plt
import pylab as pl
import matplotlib.dates as mdates
import urllib2 as ul
from datetime import datetime
from pandas.stats.moments import ewma as ewma
%matplotlib inline
```

```python
#create a variable for the web address
url = 'http://hopey.netfonds.no/tradedump.php?date=20141013&paper=NFLX.O&csv_format=txt'
open('data/NFLXTicker.csv','wb').write(ul.urlopen(url).read())
```

```python
NFLXFrame = pd.read_csv('data/NFLXTicker.csv', delimiter="\t")
```

```python
#drop NaN columns
NFLXFrame.drop(['board','source','buyer','seller','initiator'], inplace = True, axis=1)
```

```python
#convert time values to a standard python format
NFLXFrame['timeConvert'] = [datetime.strptime(t, "%Y%m%dT%H%M%S") for t in NFLXFrame['time']]
NFLXFrame['xTime'] = mdates.date2num(NFLXFrame['timeConvert'])
ind = NFLXFrame['timeConvert']
price = NFLXFrame['price']
quantity = NFLXFrame['quantity']
```

```python
#inspect dataFrame
NFLXFrame.head()
```

```python
#export for minitab (just in case)
open('data/NFLXFrame.csv','wb').write(NFLXFrame.to_csv());
```

```python
#create a time series to slice data by time windows
NFLXTSprice = pd.Series(index = ind)
NFLXTSprice[0:] = price
NFLXTSprice.head()
```

```python
#create a time series to slice data by time windows
NFLXTSquant = pd.Series(index = ind)
NFLXTSquant[0:] = quantity
NFLXTSquant.head()
```

```python
#create a datetime indices for the weighted time intervals
intervals = pd.date_range('10/13/2014', periods=270, freq='5min')
intervals2 = pd.Series(intervals,index =intervals)
indexForWeights = intervals2['2014-10-13 15:30:00':'2014-10-13 22:00:00']
index = np.arange(79)
reindex = indexForWeights.reindex(index)
reindex[0:] = indexForWeights
```

```python
#function to create five minute weighted windows
def weighted_fiveMinInt(quantityTS,priceTS,tsOfIntervals):
    weights = quantityTS*priceTS
    fiveMinIntervals = np.zeros(len(tsOfIntervals)-1)
    for i in range(len(tsOfIntervals)-1):
        weightSlice = sum(weights[tsOfIntervals[i]:tsOfIntervals[i+1]])
        quantSlice = sum(quantityTS[tsOfIntervals[i]:tsOfIntervals[i+1]])
        newPrice = weightSlice/quantSlice
        fiveMinIntervals[i] = newPrice
    return fiveMinIntervals
```

```python
weightedWindows = weighted_fiveMinInt(NFLXTSquant,NFLXTSprice,reindex)
weightedWindowsTS = pd.Series(weightedWindows, index =indexForWeights[1:])
```

```python
#open a figure window and color it white
figure = plt.figure(figsize=(20,8))

#built-in DataFrame function to plot time as x and theta as y, with custom y limits, line width, and color
TSPlot = NFLXTSprice.plot(linewidth=5, c='black')

Line2 = weightedWindowsTS.plot(linewidth=5, c='red')

#set x and y labels, title, and adjust thier sizes according
TSPlot.set_ylabel('Stock Price ($)', fontsize=16)
TSPlot.set_xlabel('Time (hour:min:sec)',fontsize=16)
TSPlot.set_title('Netflix Stock (NFLX) Ticker data from 10/13/2014',fontsize=20)
plt.tick_params(axis='both', labelsize=15)   #change the font size of the axes ticks

TSPlot.legend(('Ticker Data (Price vs. Time)','5 Min Price Intervals (weighted by quantity)'), loc=1);
```

Because of how many data points we have, it's hard to really tell the differences here. We'll have to look at a smaller time window.

```python
#MAPE Function
def MAPE(obs, pred): 
    return np.mean(np.abs((obs - pred) / obs)) * 100
```

Here are the EWMA's vs. 5min window.

```python
weightedfiveEWMA = ewma(weightedWindowsTS, span=.3333)
weightedfiftyEWMA = ewma(weightedWindowsTS, span=(float(2)/51))
```

```python
#open a figure window and color it white
figure = plt.figure(figsize=(20,8))

#built-in DataFrame function to plot time as x and theta as y, with custom y limits, line width, and color
TSPlot = weightedWindowsTS.plot(linewidth=4, c='red')

Line3 = weightedfiftyEWMA.plot(linewidth=2, c='green')
Line4 = weightedfiveEWMA.plot(linewidth=2, c='blue')

#set x and y labels, title, and adjust thier sizes according
TSPlot.set_ylabel('Stock Price ($)', fontsize=16)
TSPlot.set_xlabel('Time(hour:min:sec)',fontsize=16)
TSPlot.set_title('Netflix Stock (NFLX) Ticker data from 10/13/2014',fontsize=20)
plt.tick_params(axis='both', labelsize=15)

TSPlot.legend(('5 Min Price Intervals (weighted by quantity)','EWMA, N=50','EWMA, N=5'), loc=1);
```

```python
print("MAPE of 5min intervals vs. EWMA with N=5:",MAPE(weightedWindowsTS,weightedfiveEWMA))
print("MAPE of 5min intervals vs. EWMA with N=50:",MAPE(weightedWindowsTS,weightedfiftyEWMA))
```

The plot, and the MAPE values, indicate that EWMA with N=5 is similar to the 5 minute weighted windows that I created, while EWMA with N=50 is very eradict.

The optimized EWMA has a MAPE of .0358. The optimized EWMA appears to fit the 5 minute interval data almost perfectly, with N=5 functioning as a good approximation and N=50 not being apparently useful.

For starters, we are treating the errors in this data as iid ~N(0,simga^2).

The predicition is the last value of the first order exponential smoother (forecast origin). This is because we are treating the data as a constant process. Each subsequent prediction will also be the last value of the first order exponential smoother.

The prediciton intervals are generated by adding and subtracting a margin of error from the forcast origin, where the MOE is the product a two-sided critical z-score (alpha/2) & the estimated standard deviation of the errors.

<img src =  'images/416HW2Prob2a.PNG'>

<img src="images/miniPredProb2.PNG">

We're still assuming that the errors are iid~N(0,sigma^2).

The predicition has changed from the previous problem because we are now accounting for a trend in the data. Adding the trend term to the smoother enables us to achieve forecast values that differ from the forecast origin. In this particular case, the negative trend in the data appears to have influenced a negative trend in the predicted values.

The predicition interval is now generated by adding the trend to the forecast origin value and then adding and subtracting the MOE. It's important to note that the trend value here is a function of the forecast step. For each step beyond the forecast origin, the trend value changes. Additionally, the MOE error term is now the product of the MOE term described in 1d and a value that grows with each forecast step. So, now the predicition intervals will adjust the trend for subsequent forecast steps and the MOE will grow indefinitely.

<img src =  'images/416HW2Prob2b.PNG'>

<img src="images/miniPredProb2b.PNG">

The optimal value is a much closer fit for the data than part a, however, these optimal values greatly change the predicition interval. The lower gammma value initially creates a smaller PI, which doesn't last since gamma grows with each forecast step. Since the prediction interval has the product of alpha^2 and gamma in the MOE, the higher alpha value from the optimization causes the prediction intervals to grow much faster.

<img src = 'images/416HW2Prob3.PNG'>

<img src="images/miniPredProb3alt.PNG">

Like the previous problems, this prediction is based on the forecast origin. In this particular forecast, we are now adding the trend and the seasonal component. Since the seasonal component is a dominant cycle observed in the data, it enables the predictions to utilize the observed cycle to make a better predicition.

The prediction intervals, from what I can tell between the book and the internet are the same as the Holt's Method PI's. The PI should continue to grow with each forecast step and larger values of gamma/alpha will cause the PI's to be wider (initially) and grow faster (than the growth observed in the data above).

The Durbin-Watson test looks for autocorrelation at lag one.<br>
Ho: No Serial Autocorrelation (at lag 1)<br>
Ha: Serial Autocorrelation<br>
According to the text, for uncorrelated values the Durbin-Watson test statistic should be approximately 2. With a Durbin-Watson test statistic of 2.06 we fail to reject the Null Hypothesis. Therefore, we have no evidence of serial autocorrelation (at one lag) in our data.

<img src =  'images/416HW2Prob4e.PNG'>

<img src="images/lbq1.PNG">

The LBQ Test looks for autocorrelation in the the lags from 1 to p. In this case, p = 75.<br>
Ho: No Serial Autocorrelation (from lags 1 to 75)<br>
Ha: Serial Autocorrelation<br>
With a LBQ Test statistics of 70.028 and a p-value of .6408 (using a Chi-square distribution with 75 degrees of freedom) we fail to reject the Null Hypothesis. Therefore, we have no evidence of serial autocorrelation (from lag 1 to 75) in our data.

<img src = 'images/416HW2Prob4f.PNG'>

Yes, the results were consistent. Each method indicated that there was no apparent autocorrelation at lag 1. Additionally, the correlogram and the LBQ test indicated that there was no autocorrelation at from lags 1 to 75 (with the possible exception of lag 10 in the correlogram).
