# Simple Moving Average (SMA)

```python
def SMA(vec, n):
    sma = np.zeros([len(vec),1])  #created an empty array to store values
    for i in range(n, len(vec)):    #look at values from n to length
        window = vec[i-n:i]       #taking average over these values
        sma[i] = (np.mean(window))    #store average in array
    sma[:n] = None                    #fill first n values with None
    return sma
```

```python
sliceFrame['SMA'] = SMA(sliceFrame.Detections,24)
```

# Centered Moving Average (CMA)

```python
def CMA(vec, n):
    c = n/2
    cma = np.zeros([len(vec),1])  #created an empty array to store values
    for i in range(c, len(vec)-c):    #look at values from n/2 to length-n/2
        window = vec[i-c:i+c+1]       #taking average over these values
        cma[i] = (np.mean(window))    #store average in array
    cma[:c] = None                    #fill first n/2 values with None
    cma[(len(vec)-c):] = None         #fill last n/2 values with None
    return cma
```

```python
sliceFrame['CMA'] = CMA(sliceFrame.Detections,24)
```

# A Simple Weighted Moving Average (SWMA)

```python
def SWMA(vec, n):
    weights = np.arange(n+2)     #generate weights    
    sWeights = weights[1:]   #sliced weights
    swma = np.zeros([len(vec),1])  #created an empty array to store values
    for i in range(n, len(vec)):    #look at values from n to length
        window = vec[i-n:i+1]       #taking average over these values
        weightedWindow=window*sWeights  
        swma[i] = (2*(np.sum(weightedWindow))/((n+1)*(n+2)))    #store average in array
    swma[:n] = None                    #fill first n values with None
    return swma
```

```python
sliceFrame['SWMA'] = SWMA(sliceFrame.Detections,24)
```

# First-Order Exponential Smoothing

```python
def EWMA(vec,theta):
    n = len(vec)
    if n==1:
        return vec[0]
    else:
        return (1-theta)*vec[n] +((1-theta)**)*vec[n-1]
```

```python
sliceFrame['EWMA'] = EWMA(sliceFrame.Detections,20)
```
