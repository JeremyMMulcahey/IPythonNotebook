```python
#create a variable for the web address
url = 'http://hopey.netfonds.no/tradedump.php?date=20141013&paper=NFLX.O&csv_format=txt'
open('data/NFLXTicker.csv','wb').write(ul.urlopen(url).read())
```

```python
NFLXFrame = pd.read_csv('data/NFLXTicker.csv', delimiter="\t")
```

```python
#drop NaN columns
NFLXFrame.drop(['board','source','buyer','seller','initiator'], inplace = True, axis=1)
```

```python
#convert time values to a standard python format
NFLXFrame['timeConvert'] = [datetime.strptime(t, "%Y%m%dT%H%M%S") for t in NFLXFrame['time']]
NFLXFrame['xTime'] = mdates.date2num(NFLXFrame['timeConvert'])
ind = NFLXFrame['timeConvert']
price = NFLXFrame['price']
quantity = NFLXFrame['quantity']
```

```python
#inspect dataFrame
NFLXFrame.head()
```

```python
#export for minitab (just in case)
open('data/NFLXFrame.csv','wb').write(NFLXFrame.to_csv());
```

```python
#create a time series to slice data by time windows
NFLXTSprice = pd.Series(index = ind)
NFLXTSprice[0:] = price
NFLXTSprice.head()
```

```python
#create a time series to slice data by time windows
NFLXTSquant = pd.Series(index = ind)
NFLXTSquant[0:] = quantity
NFLXTSquant.head()
```

```python
#create a datetime indices for the weighted time intervals
intervals = pd.date_range('10/13/2014', periods=270, freq='5min')
intervals2 = pd.Series(intervals,index =intervals)
indexForWeights = intervals2['2014-10-13 15:30:00':'2014-10-13 22:00:00']
index = np.arange(79)
reindex = indexForWeights.reindex(index)
reindex[0:] = indexForWeights
```

```python
#function to create five minute weighted windows
def weighted_fiveMinInt(quantityTS,priceTS,tsOfIntervals):
    weights = quantityTS*priceTS
    fiveMinIntervals = np.zeros(len(tsOfIntervals)-1)
    for i in range(len(tsOfIntervals)-1):
        weightSlice = sum(weights[tsOfIntervals[i]:tsOfIntervals[i+1]])
        quantSlice = sum(quantityTS[tsOfIntervals[i]:tsOfIntervals[i+1]])
        newPrice = weightSlice/quantSlice
        fiveMinIntervals[i] = newPrice
    return fiveMinIntervals
```

```python
weightedWindows = weighted_fiveMinInt(NFLXTSquant,NFLXTSprice,reindex)
weightedWindowsTS = pd.Series(weightedWindows, index =indexForWeights[1:])

#minitab export
weightedWindowsTS.to_csv('data/5min.csv')
```

```python
#open a figure window and color it white
figure = plt.figure(figsize=(20,8))

#built-in DataFrame function to plot time as x and theta as y, with custom y limits, line width, and color
TSPlot = NFLXTSprice.plot(linewidth=5, c='black')

Line2 = weightedWindowsTS.plot(linewidth=5, c='red')

#set x and y labels, title, and adjust thier sizes according
TSPlot.set_ylabel('Stock Price ($)', fontsize=16)
TSPlot.set_xlabel('Time (hour:min:sec)',fontsize=16)
TSPlot.set_title('Netflix Stock (NFLX) Ticker data from 10/13/2014',fontsize=20)
plt.tick_params(axis='both', labelsize=15)   #change the font size of the axes ticks

TSPlot.legend(('Ticker Data (Price vs. Time)','5 Min Price Intervals (weighted by quantity)'), loc=1);
```

```python
#MAPE Function
def MAPE(obs, pred): 
    return np.mean(np.abs((obs - pred) / obs)) * 100
```

```python
#online fuction from Connor Johnson
def double_ewma( x, alpha, beta ):
    N = x.size
    s = np.zeros(( N, ))
    b = np.zeros(( N, ))
    s[0] = x[0]
    for i in range( 1, N ):
        s[i] = alpha * x[i] + ( 1 - alpha )*( s[i-1] + b[i-1] )
        b[i] = beta * ( s[i] - s[i-1] ) + ( 1 - beta ) * b[i-1]
    return s
```

```python
doubleEWMA = double_ewma(weightedWindowsTS, .2,.2)
```

```python
doubleEWMATS = pd.Series(doubleEWMA, index =indexForWeights[1:])
```
