```python
from IPython.display import Image
import numpy as np
import pandas as pd
import math
import matplotlib
import matplotlib.pyplot as plt
import pylab as pl
import matplotlib.dates as mdates
import urllib2 as ul
from datetime import datetime
from pandas.stats.moments import ewma as ewma
%matplotlib inline
```

```python
#create a variable for the web address
url = 'http://hopey.netfonds.no/tradedump.php?date=20141013&paper=NFLX.O&csv_format=txt'
open('data/NFLXTicker.csv','wb').write(ul.urlopen(url).read())
```

```python
NFLXFrame = pd.read_csv('data/NFLXTicker.csv', delimiter="\t")
```

```python
#drop NaN columns
NFLXFrame.drop(['board','source','buyer','seller','initiator'], inplace = True, axis=1)
```

```python
#convert time values to a standard python format
NFLXFrame['timeConvert'] = [datetime.strptime(t, "%Y%m%dT%H%M%S") for t in NFLXFrame['time']]
NFLXFrame['xTime'] = mdates.date2num(NFLXFrame['timeConvert'])
ind = NFLXFrame['timeConvert']
price = NFLXFrame['price']
quantity = NFLXFrame['quantity']
```

```python
#inspect dataFrame
NFLXFrame.head()
```

```python
#create a time series to slice data by time windows
NFLXTSprice = pd.Series(index = ind)
NFLXTSprice[0:] = price
NFLXTSprice.head()
```

```python
#create a time series to slice data by time windows
NFLXTSquant = pd.Series(index = ind)
NFLXTSquant[0:] = quantity
NFLXTSquant.head()
```

```python
#create a datetime indices for the weighted time intervals
intervals = pd.date_range('10/13/2014', periods=270, freq='5min')
intervals2 = pd.Series(intervals,index =intervals)
indexForWeights = intervals2['2014-10-13 15:30:00':'2014-10-13 22:00:00']
index = np.arange(79)
reindex = indexForWeights.reindex(index)
reindex[0:] = indexForWeights
```

```python
#function to create five minute weighted windows
def weighted_fiveMinInt(quantityTS,priceTS,tsOfIntervals):
    weights = quantityTS*priceTS
    fiveMinIntervals = np.zeros(len(tsOfIntervals)-1)
    for i in range(len(tsOfIntervals)-1):
        weightSlice = sum(weights[tsOfIntervals[i]:tsOfIntervals[i+1]])
        quantSlice = sum(quantityTS[tsOfIntervals[i]:tsOfIntervals[i+1]])
        newPrice = weightSlice/quantSlice
        fiveMinIntervals[i] = newPrice
    return fiveMinIntervals
```

```python
weightedWindows = weighted_fiveMinInt(NFLXTSquant,NFLXTSprice,reindex)
weightedWindowsTS = pd.Series(weightedWindows, index =indexForWeights[1:])
```

```python
#open a figure window and color it white
figure = plt.figure(figsize=(20,8))

#built-in DataFrame function to plot time as x and theta as y, with custom y limits, line width, and color
TSPlot = NFLXTSprice.plot(linewidth=5, c='black')

Line2 = weightedWindowsTS.plot(linewidth=5, c='red')

#set x and y labels, title, and adjust thier sizes according
TSPlot.set_ylabel('Stock Price', fontsize=16)
TSPlot.set_xlabel('Time',fontsize=16)
TSPlot.set_title('Netflix Stock (NFLX) Ticker data from 10/13/2014',fontsize=20)
plt.tick_params(axis='both', labelsize=15)   #change the font size of the axes ticks

TSPlot.legend(('Ticker Data (Price vs. Time)','5 Min Price Intervals (weighted by quantity)'), loc=1);
```

```python
fiveEWMA = ewma(NFLXTSprice, span=.3333)
fiftyEWMA = ewma(NFLXTSprice, span=(float(2)/51))
```

```python
#open a figure window and color it white
figure = plt.figure(figsize=(20,8))

#built-in DataFrame function to plot time as x and theta as y, with custom y limits, line width, and color
TSPlot = NFLXTSprice.plot(linewidth=10, c='yellow')

Line3 = fiftyEWMA.plot(linewidth=2, c='black')
Line4 = fiveEWMA.plot(linewidth=.5, c='blue')
Line2 = weightedWindowsTS.plot(linewidth=4, c='red')

#set x and y labels, title, and adjust thier sizes according
TSPlot.set_ylabel('Stock Price', fontsize=16)
TSPlot.set_xlabel('Time',fontsize=16)
TSPlot.set_title('Netflix Stock (NFLX) Ticker data from 10/13/2014',fontsize=20)
plt.tick_params(axis='both', labelsize=15)

TSPlot.legend(('Ticker Data (Price vs. Time)','EWMA, N=50','EWMA, N=5','5 Min Price Intervals (weighted by quantity)'), loc=1);
```

Because of how many data points we have, it's hard to really tell the differences here. We'll have to look at a smaller time window.

```python
zoomPrice = NFLXTSprice['2014-10-13 16:00:00':'2014-10-13 17:00:00']
zoomWeight = weightedWindowsTS['2014-10-13 16:00:00':'2014-10-13 17:00:00']
zoom5EWMA = fiveEWMA['2014-10-13 16:00:00':'2014-10-13 17:00:00']
zoom50EWMA = fiftyEWMA['2014-10-13 16:00:00':'2014-10-13 17:00:00']
```

```python
#open a figure window and color it white
figure = plt.figure(figsize=(20,8))

#built-in DataFrame function to plot time as x and theta as y, with custom y limits, line width, and color
TSPlot = zoomPrice.plot(linewidth=8, c='black')

Line3 = zoom50EWMA.plot(linewidth=2, c='red')
Line4 = zoom5EWMA.plot(linewidth=2, c='blue')
Line2 = zoomWeight.plot(linewidth=2, c='green')

#set x and y labels, title, and adjust thier sizes according
TSPlot.set_ylabel('Stock Price', fontsize=16)
TSPlot.set_xlabel('Time',fontsize=16)
TSPlot.set_title('Netflix Stock (NFLX) Ticker data from 10/13/2014',fontsize=20)
plt.tick_params(axis='both', labelsize=15)

TSPlot.legend(('Ticker Data (Price vs. Time)','EWMA, N=50','EWMA, N=5','5 Min Price Intervals (weighted by quantity)'), loc=1);
```

```python
#MAPE Function
def MAPE(obs, pred): 
    return np.mean(np.abs((obs - pred) / obs)) * 100
```

```python
MAPE(NFLXTSprice,fiveEWMA)
```

```python
MAPE(NFLXTSprice,fiftyEWMA)
```

The EWMA estimate with alpha = .3333 (N=5) has a very similar fit to the original data. It's difficut to tell them apart. The EWMA estimate with alpha = .039 (N=50) is a bit eradict. In the plot it fluctuates about the plots of the data and EWMA estimate of N=5. The MAPE values indicate that EWMA with N=5 is a closer fit to the data than EWMA with N=50.

```python
#SSE for TS
sum(NFLXTSprice-fiveEWMA)
```

```python
#function to optimize SSE
def optimize_SSE(ts,lowerBound,upperBound,precision):
    alpha = np.arange(lowerBound,upperBound,precision)
    optimized = 0
    SSEcheck = 10000000
    
    #run an ewma for each alpha and see if it lowered SSE
    for i in alpha:
        ewma1 = ewma(ts, span=i)
        SSE = abs(sum(ts-ewma1))
        if SSE < SSEcheck:
            optimized = i
            SSEcheck = SSE
            
    #after all alphas are checked, see if a step outside the upperBound
    #lowers SSE then prints the next upperbound to try if an optimal alpha
    #does not exist within the original specified bounds
    ewma2 = ewma(ts,span=(upperBound+.001))
    SSE2 = abs(sum(ts-ewma2))
    if SSE2 < SSEcheck:
        nextUB = upperBound+1
        print(SSE, SSE2)
        print('Run upperBound =',nextUB)
        
    #if optimal alpha did exist witin the range, the original bounds
    #return it and print the values
    else:
        print("SSE:",SSEcheck)
        print("Alpha:",optimized)
        return optimized
```

```python
#note: I ran this checking alpha from 0-2, changed it to a smaller window
#of (.8,1.2) so I could rerun it (since it's a long calculation)
optAlpha = optimize_SSE(NFLXTSprice,0.8,1.2,.0001)
optEwma = ewma(NFLXTSprice,span=optAlpha)
```

```python
#open a figure window and color it white
figure = plt.figure(figsize=(20,8))

#built-in DataFrame function to plot time as x and theta as y, with custom y limits, line width, and color
TSPlot = NFLXTSprice.plot(linewidth=10, c='yellow')

Line3 = optEwma.plot(linewidth=2, c='black')
Line2 = weightedWindowsTS.plot(linewidth=4, c='red')

#set x and y labels, title, and adjust thier sizes according
TSPlot.set_ylabel('Stock Price', fontsize=16)
TSPlot.set_xlabel('Time',fontsize=16)
TSPlot.set_title('Netflix Stock (NFLX) Ticker data from 10/13/2014',fontsize=20)
plt.tick_params(axis='both', labelsize=15)

TSPlot.legend(('Ticker Data (Price vs. Time)','EWMA, Alpha=.99999','5 Min Price Intervals (weighted by quantity)'), loc=1);
```

```python
print(MAPE(NFLXTSprice,fiveEWMA))
print(MAPE(NFLXTSprice,fiftyEWMA))
print(MAPE(NFLXTSprice,optEwma))
```
