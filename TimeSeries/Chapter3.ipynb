# Chapter 3<br> ARMA Models

```python
import numpy as np
import pandas as pd
import math
import matplotlib
import matplotlib.pyplot as plt
import statsmodels.tsa as tsa
from statsmodels.tsa.arima_model import ARIMA as arima
import datetime
import time
%matplotlib inline
```

<img src = "images/HW3Prob1.PNG">

```python
#read the csv file
PendData = np.loadtxt('data/pendData.csv', skiprows=1, delimiter=',')

#convert the numpy structure into a pandas DataFrame
PendFrame = pd.DataFrame(PendData,columns=['time(sec)','position(theta)','AR2','ARstep','ARstep2','ARstep3','ARstep4'])
AR2 = PendFrame['AR2']
Time = PendFrame['time(sec)']
Theta = PendFrame['position(theta)']
ARstep = PendFrame['ARstep']
ARstep2 = PendFrame['ARstep2']
ARstep3 = PendFrame['ARstep3']
ARstep4 = PendFrame['ARstep4']
```

```python
PendTS = pd.Series(index = Time)
PendTS[0:] = Theta
```

```python
AR2TS = pd.Series(index = Time)
AR2TS[0:] = AR2
```

```python
#open a figure window and color it white
figure = plt.figure(figsize=(20,8))

TSPlot = PendTS.plot(linewidth=2, c='black', ylim=(-26,30))

Line2 = AR2TS.plot(linewidth=1, c='red')

#set x and y labels, title, and adjust thier sizes according
TSPlot.set_ylabel('position(theta)', fontsize=20)
TSPlot.set_xlabel('Time (seconds)',fontsize=20)
TSPlot.set_title('Pendulum Data fit with AR2 Model',fontsize=20)
plt.tick_params(axis='both', labelsize=15) 
plt.scatter(PendFrame['time(sec)'],PendFrame['position(theta)'], s=10, c='black')


TSPlot.legend(('Pendulum Fit', 'AR2 Model','Pendulum Points'), loc=1);
```

<img src = "images/HW3Prob1b.PNG">

```python
nume = sum((PendTS - AR2TS)**2)
denom = sum((PendTS-np.mean(PendTS))**2)
1 - (float(nume)/denom)
```

Approximately 98% of the variation in the pendulum time series data (theta) can be explained by an optimized AR2 model of theta.

<img src = "images/HW3Prob1c.PNG">

```python
ARstepTS = pd.Series(index = Time)
ARstepTS[0:] = ARstep
```

```python
#open a figure window and color it white
figure = plt.figure(figsize=(20,8))

TSPlot = PendTS.plot(linewidth=2, c='black', ylim=(-26,30))

Line2 = ARstepTS.plot(linewidth=1, c='red')

#set x and y labels, title, and adjust thier sizes according
TSPlot.set_ylabel('position(theta)', fontsize=20)
TSPlot.set_xlabel('Time (seconds)',fontsize=20)
TSPlot.set_title('Pendulum Data fit with AR2step Model',fontsize=20)
plt.tick_params(axis='both', labelsize=15) 
plt.scatter(PendFrame['time(sec)'],PendFrame['position(theta)'], s=10, c='black')


TSPlot.legend(('Pendulum Fit', 'AR2step Model','Pendulum Points'), loc=1);
```

<img src = "images/HW3Prob1d.PNG">

```python
nume = sum((PendTS - ARstepTS)**2)
denom = sum((PendTS-np.mean(PendTS))**2)
1 - (float(nume)/denom)
```

Comparing the R-square values (.98 vs. .70) we can see that the 1-step ahead did not match the pendulum data as well as the AR(2) model. We can also see this in the time series plot from part c). The first peak in the 1-step model is quite large and the rest of the plot seems to be shifted (slightly) away from the sine wave, where as the AR(2) model appeared to overlay on the sine wave almost point for point (with the exception of the 3 or 4 overestimated peaks).

```python
#if the coefficients are both large
ARstepTS2 = pd.Series(index = Time)
ARstepTS2[0:] = ARstep2

#if the coefficients are both small
ARstepTS3 = pd.Series(index = Time)
ARstepTS3[0:] = ARstep3

#if the coefficients have their signs switched
ARstepTS4 = pd.Series(index = Time)
ARstepTS4[0:] = ARstep4
```

Let's look at the large case first. If both coefficients are increased to 5 and -5 (from 1.7 and -1).

```python
#open a figure window and color it white
figure = plt.figure(figsize=(20,8))

TSPlot = PendTS.plot(linewidth=2, c='black', ylim=(-26,30))

Line2 = ARstepTS2.plot(linewidth=1, c='red')

#set x and y labels, title, and adjust thier sizes according
TSPlot.set_ylabel('position(theta)', fontsize=20)
TSPlot.set_xlabel('Time (seconds)',fontsize=20)
TSPlot.set_title('Pendulum Data fit with AR2 Model (Large Coefficients)',fontsize=20)
plt.tick_params(axis='both', labelsize=15) 
plt.scatter(PendFrame['time(sec)'],PendFrame['position(theta)'], s=10, c='black')


TSPlot.legend(('Pendulum Fit', 'AR2 Model','Pendulum Points'), loc=1);
```

Increasing the coefficients greatly scales the fit of the AR(2) model.

What happens when the cofficients are reduced? Now .3 and -.3, respectively.

```python
#open a figure window and color it white
figure = plt.figure(figsize=(20,8))

TSPlot = PendTS.plot(linewidth=2, c='black', ylim=(-26,30))

Line2 = ARstepTS3.plot(linewidth=1, c='red')

#set x and y labels, title, and adjust thier sizes according
TSPlot.set_ylabel('position(theta)', fontsize=20)
TSPlot.set_xlabel('Time (seconds)',fontsize=20)
TSPlot.set_title('Pendulum Data fit with AR2 Model (Small Coefficients)',fontsize=20)
plt.tick_params(axis='both', labelsize=15) 
plt.scatter(PendFrame['time(sec)'],PendFrame['position(theta)'], s=10, c='black')


TSPlot.legend(('Pendulum Fit', 'AR2 Model','Pendulum Points'), loc=1);
```

Smaller scalers shrinks the fit and appears to have shifted the peaks.

Next we'll switch the signs of phi_1 and phi_2.

```python
#open a figure window and color it white
figure = plt.figure(figsize=(20,8))

TSPlot = PendTS.plot(linewidth=2, c='black', ylim=(-26,30))

Line2 = ARstepTS4.plot(linewidth=1, c='red')

#set x and y labels, title, and adjust thier sizes according
TSPlot.set_ylabel('position(theta)', fontsize=20)
TSPlot.set_xlabel('Time (seconds)',fontsize=20)
TSPlot.set_title('Pendulum Data fit with AR2 Model (Reversed Signs)',fontsize=20)
plt.tick_params(axis='both', labelsize=15) 
plt.scatter(PendFrame['time(sec)'],PendFrame['position(theta)'], s=10, c='black')


TSPlot.legend(('Pendulum Fit', 'AR2 Model','Pendulum Points'), loc=1);
```

This resulted in another phase shift and some scattered peaks. This one was a bit unexpected compared to the previous results.
